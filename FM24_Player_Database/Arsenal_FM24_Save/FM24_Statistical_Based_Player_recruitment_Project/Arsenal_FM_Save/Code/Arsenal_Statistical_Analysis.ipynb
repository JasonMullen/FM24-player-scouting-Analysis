```python
import pandas as pd

# Define a function to calculate the defensive score for a player
def defensive_score(row):
    # Define weights for each metric
    weights = {
        'Saves Parried': 0.15,
        'Saves Tipped': 0.15,
        'Saves Held': 0.2,
        'Save Percentage': 0.1,
        'Inteceptions per 90': 0.1,
        'Expected Save Percentage': 0.1,
        'Games Win Ratio': 0.1,
        'Total Saves Per 90': 0.05,
        'Key Passes Per 90': 0.05,
        'Pass Completion Ratio': 0.05,
        'Pens Saved Ratio': 0.05
    }
    
    # Extract relevant values from the row
    relevant_values = row[['Saves Parried', 'Saves Tipped', 'Saves Held', 
                           'Save Percentage', 'Inteceptions per 90', 
                           'Expected Save Percentage', 'Games Win Ratio', 
                           'Total Saves Per 90', 'Key Passes Per 90', 
                           'Pass Completion Ratio', 'Pens Saved Ratio']]
    
    # Calculate weighted sum
    defensive_weighted_sum = sum(value * weights[column] for column, value in relevant_values.items())
    
    return defensive_weighted_sum

# Define a function to calculate the offensive score for a player
def offensive_score(row):
    # Define weights for each metric
    weights = {
        'Passes Completed per 90': 0.15,
        'Pass Completion Ratio': 0.2,
        'Key Passes Per 90': 0.25,
        'Passes attempted per 90': 0.2,
        'Progressive Passes Per 90': 0.2
    }
    
    # Extract relevant values from the row
    relevant_values = row[['Passes Completed per 90', 'Pass Completion Ratio', 
                           'Key Passes Per 90', 'Passes attempted per 90', 
                           'Progressive Passes Per 90']]
    
    # Calculate weighted sum
    offensive_weighted_sum = sum(value * weights[column] for column, value in relevant_values.items())
    
    return offensive_weighted_sum

# Define a function to calculate the rating score for a player
def rating_score(row):
    # Define weights for each metric
    weights = {
        'Average Rating': 0.6,
        'Games Win Ratio': 0.3,
        'Player Of The Match': 0.1
    }
    
    # Extract relevant values from the row
    relevant_values = row[['Average Rating', 'Games Win Ratio', 'Player Of The Match']]
    
    # Calculate weighted sum
    rating_score_weighted_sum = sum(value * weights[column] for column, value in relevant_values.items())
    
    return rating_score_weighted_sum

# Load your DataFrame containing the data


# Convert relevant columns to numeric data types
numeric_columns = ['Saves Parried', 'Saves Tipped', 'Saves Held', 'Save Percentage', 
                   'Inteceptions per 90', 'Expected Save Percentage', 'Games Win Ratio', 
                   'Total Saves Per 90', 'Key Passes Per 90', 'Pass Completion Ratio', 'Pens Saved Ratio']

sgk_df[numeric_columns] = sgk_df[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Calculate scores for each player
defensive_scores = sgk_df.apply(defensive_score, axis=1)
offensive_scores = sgk_df.apply(offensive_score, axis=1)
rating_scores = sgk_df.apply(rating_score, axis=1)

# Concatenate the scores with the original DataFrame
result_df = pd.concat([sgk_df[['Name', 'Club', 'Age', 'Average Transfer Value']], 
                       defensive_scores.rename('Defensive Score'), 
                       offensive_scores.rename('Offensive Score'), 
                       rating_scores.rename('Rating Score')], axis=1)

# Display the resulting DataFrame
result_df
```

```python
# Calculate Sweeper Keeper score
result_df['Sweeper Keeper Score'] = (result_df['Defensive Score'] + result_df['Offensive Score'] + result_df['Rating Score']) / 3

# Sort the final DataFrame by 'Sweeper Keeper Score' and 'Defensive Score' in descending order
final_result_df = result_df.sort_values(by=['Sweeper Keeper Score', 'Defensive Score'], ascending=False)

# Display the resulting DataFrame
final_result_df.head(100)
```

```python
# Sort the DataFrame by each metric
defensive_report = result_df.sort_values(by='Defensive Score', ascending=False).head(20)
offensive_report = result_df.sort_values(by='Offensive Score', ascending=False).head(20)
rating_report = result_df.sort_values(by='Rating Score', ascending=False).head(20)
sweeper_keeper_report = result_df.sort_values(by='Sweeper Keeper Score', ascending=False).head(20)

# Define a function to generate HTML report and save it to a file
def generate_html_report(report, title, filename):
    html_content = report[['Name', 'Club', 'Age', 'Average Transfer Value', 'Defensive Score', 'Offensive Score', 'Rating Score', 'Sweeper Keeper Score']].to_html(index=False)
    try:
        with open(filename, 'w') as file:
            file.write(f"<h1>{title}</h1>")
            file.write(html_content)
        print(f"Report '{title}' was generated successfully and saved as '{filename}'.")
    except Exception as e:
        print(f"Failed to generate report '{title}': {str(e)}")

# Generate and save HTML reports
generate_html_report(defensive_report, "Top 20 Defensive Keepers", "defensive_report.html")
generate_html_report(offensive_report, "Top 20 Offensive Sweeper Keepers", "offensive_report.html")
generate_html_report(rating_report, "Top 20 Rating Scores", "rating_report.html")
generate_html_report(sweeper_keeper_report, "Top 20 Sweeper Keeper Scores", "sweeper_keeper_report.html")
```
