__Goal Keeper attribute algorithm__

    This section of code takes in all the the attributes attributed to a sweeper keeper and calculates a final score.
    The higher the score the better the player and the better the player the more he'll be able to fulfil his role in your tactic.

```python
#Formula: (gk_e + gk_c + gk_s) / Total_score
sweeper_df['gk_essential'] = ((sweeper_df['Agility'] + sweeper_df['Reflexes']) * 5)




sweeper_df['gk_core'] = ((sweeper_df['Ariel Reach'] + sweeper_df['Command Of Area'] + sweeper_df['Communication'] +
                          sweeper_df['Handling'] + sweeper_df['Kicking'] + sweeper_df['Concentration'] + 
                          sweeper_df['Positioning']) * 3)





sweeper_df['gk_secondary'] = ((sweeper_df['First Touch'] + sweeper_df['Passing'] + sweeper_df['Throwing'] +
                                sweeper_df['Vision'] + sweeper_df['Acceleration']) * 1)




sweeper_df['gk'] = ((sweeper_df['gk_essential'] + sweeper_df['gk_core'] + sweeper_df['gk_secondary']) / 36)
sweeper_df['gk'] = sweeper_df['gk'].round(1)
```

__GK Visualization__
    
    In this section of code where going to strip the Transfer Value and Salary unnessary characters and convert the floats into strings.
    From there we're going to use their transfer value and Gk score in a scatterplot. We're hoping to find who the best keeper is and how much they cost
    Along with who's value for money keeper(Highest score and lowest price)
    

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import ipywidgets as widgets
from ipywidgets import interactive

# Function to convert transfer value
def convert_transfer_value(value):
    if isinstance(value, str):
        if 'Not for Sale' in value:
            return 0  # Return 0 for "Not for Sale"
        
        # Remove unwanted characters and split range values if present
        value = value.replace('', '').replace('M', '000000').replace('K', '000')
        if '-' in value:
            lower, upper = value.split('-')
            lower_val = float(lower.strip())
            upper_val = float(upper.strip())
            value = (lower_val + upper_val) / 2
        else:
            value = float(value.strip())
        
        return value / 1000000  # Convert to millions
    else:
        return value  # Return the value unchanged if it's already a float

# Function to strip salary
def strip_salary(value):
    if isinstance(value, str):
        return float(value.replace('', '').replace(',', '').replace(' p/w', ''))
    else:
        return value  # Return the value unchanged if it's not a string

# Apply conversion functions to 'Transfer Value' and 'Salary' columns
sweeper_df['Transfer Value'] = sweeper_df['Transfer Value'].apply(convert_transfer_value)
sweeper_df['Salary'] = sweeper_df['Salary'].apply(strip_salary)

# Find the player with the highest transfer value and highest salary
highest_transfer_value = sweeper_df.loc[sweeper_df['Transfer Value'].idxmax()]
highest_salary = sweeper_df.loc[sweeper_df['Salary'].idxmax()]


# Define a function to update the plot based on gk values
def update_plot_gk(salary_range, gk_range):
    plt.figure(figsize=(20, 10))
    
    # Define conditions for color coding based on gk ranges
    conditions = [
        (sweeper_df['gk'] > 10) & (sweeper_df['gk'] <= 11),
        (sweeper_df['gk'] > 11) & (sweeper_df['gk'] <= 12),
        (sweeper_df['gk'] > 12) & (sweeper_df['gk'] <= 13),
        (sweeper_df['gk'] > 13) & (sweeper_df['gk'] <= 14),
        (sweeper_df['gk'] > 14) & (sweeper_df['gk'] <= 15),
        (sweeper_df['gk'] > 15) & (sweeper_df['gk'] <= 16),
    ]
    colors = ['orange', 'yellow', 'green', 'blue', 'purple', 'red']
    
    # Scatter plot with color coordination
    for i in range(len(conditions)):
        subset = sweeper_df[conditions[i]]
        plt.scatter(subset['Transfer Value'], subset['gk'], color=colors[i], label=f'Range {i+1}', s=100)
    
    # Annotate each point with player name
    for i, txt in enumerate(sweeper_df['Name']):
        plt.annotate(txt, (sweeper_df['Transfer Value'][i], sweeper_df['gk'][i]), fontsize=12, weight='bold')
    
    plt.xlabel('Transfer Value', fontsize=14)
    plt.ylabel('GK Score', fontsize=14)
    plt.title('Goal Keeper Affordability', fontsize=16)
    # Custom legend labels
    legend_labels = [
        '11 < gk <= 12',
        '12 < gk <= 13',
        '13 < gk <= 14',
        '14 < gk <= 15',
        '15 < gk <= 16',
    ]
    plt.legend(legend_labels, fontsize=12)
    plt.grid(True)
    plt.xlim(salary_range[0], salary_range[1])  # Set x-axis range based on salary range selection
    plt.ylim(10, gk_range[1])  # Set the y-axis limit to start from 10 and end at the selected gk range
    plt.show()

# Create interactive sliders with range selection capability
salary_slider_gk = widgets.FloatRangeSlider(value=[sweeper_df['Transfer Value'].min(), sweeper_df['Transfer Value'].max()], min=sweeper_df['Transfer Value'].min(), max=sweeper_df['Transfer Value'].max(), step=0.1, description='Transfer Value Range')
gk_slider = widgets.FloatRangeSlider(value=[10, sweeper_df['gk'].max()], min=10, max=sweeper_df['gk'].max() + 5, step=0.1, description='gk Range')

# Create interactive plot
interactive_plot_gk = interactive(update_plot_gk, salary_range=salary_slider_gk, gk_range=gk_slider)
interactive_plot_gk
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import ipywidgets as widgets
from ipywidgets import interactive
import re
import mplcursors  # Import mplcursors library for hover and click functionality
#=========================================================================================================================================================================================================================================================================================================
# Function to convert transfer value
# Function to convert transfer value
def convert_transfer_value(value):
    if isinstance(value, str):
        if 'Not for Sale' in value:
            return 0  # Return 0 for "Not for Sale"
        
        # Remove unwanted characters and split range values if present
        value = value.replace('', '').replace('M', '000000').replace('K', '000')
        if '-' in value:
            lower, upper = value.split('-')
            lower_val = float(lower.strip())
            upper_val = float(upper.strip())
            value = (lower_val + upper_val) / 2
        else:
            value = float(value.strip())
        
        return value / 1000000  # Convert to millions
    else:
        return value  # Return the value unchanged if it's already a float

# Function to strip salary
def strip_salary(value):
    if isinstance(value, str):
        # Extract digits and convert to float
        salary_digits = re.findall(r'\d+', value)
        if salary_digits:
            return float(''.join(salary_digits))
        else:
            return 0  # Return 0 if no digits are found
    else:
        return value  # Return the value unchanged if it's not a string

#=========================================================================================================================================================================================================================================================================================================
# Apply conversion functions to 'Transfer Value' and 'Salary' columns
new_fb_wcb_df['Transfer Value'] = new_fb_wcb_df['Transfer Value'].apply(convert_transfer_value)
new_fb_wcb_df['Salary'] = new_fb_wcb_df['Salary'].apply(strip_salary)
```
