```python
import pandas as pd
from IPython.display import display
import ipywidgets as widgets
pd.set_option('display.max_columns', None)  # Display all columns
pd.set_option('display.width', 1000)       # Adjust display width for better readability
pd.set_option('display.max_rows', None)    # Display all rows (if needed)



date = '/Users/JumpMan/Downloads/FM Data Analytics/Newcastle/Newcastle/Moneyball Scouting/Top 5 League Analysis/2026-2027/Summer 2026-27 Top 14 Leagues.csv'

# Detect encoding
with open(file_path, 'rb') as f:
    detected = chardet.detect(f.read(10000))  # Detect encoding in the first 10,000 bytes
    encoding = detected['encoding']

# Load CSV with detected encoding and handle BOM (if any)
try:
    data = pd.read_csv(file_path, encoding='utf-8-sig')  # 'utf-8-sig' handles BOM
    display(data.head())
except UnicodeDecodeError as e:
    print("Error loading file with utf-8-sig encoding:", e)


# Columns to keep in addition to the metrics
additional_columns = ['Name', 'Position', 'Age', 'Height', 'Weight', 'Club', 
                      'Division', 'Nationality', 'Personality', 'Media Handling', 
                      'Wage', 'Transfer Value', 'Preferred Foot', 'Minutes Played']

# Define the widget elements without the "(percentile)" filter for troubleshooting
metric_select = widgets.SelectMultiple(
    options=[col for col in data.columns if data[col].dtype in ['float64', 'int64']],
    description='Metrics',
    disabled=False,
    layout=widgets.Layout(width='50%', height='300px')
)

min_minutes_slider = widgets.IntSlider(value=500, min=0, max=3000, step=100, description='Min Minutes')
top_n_slider = widgets.IntSlider(value=5, min=1, max=20, description='Top N')

# Search input for player name
player_search = widgets.Text(description='Player Name', placeholder='Type player name here')

# Dropdown filters for categorical fields
team_dropdown = widgets.Dropdown(
    options=[''] + sorted(data['Club'].dropna().unique().tolist()),
    description='Team'
)
league_dropdown = widgets.Dropdown(
    options=[''] + sorted(data['Division'].dropna().unique().tolist()),
    description='League'
)
nationality_dropdown = widgets.Dropdown(
    options=[''] + sorted(data['Nationality'].dropna().unique().tolist()),
    description='Nationality'
)

# Dictionary to store metric input fields
metric_inputs = {}

# Function to create metric input fields
def create_metric_inputs(change):
    # Clear existing widgets in metric_inputs
    for metric, inputs in metric_inputs.items():
        inputs[0].close()
        inputs[1].close()
    metric_inputs.clear()

    # Update the VBox children with the new selected metrics
    inputs_vbox.children = []
    for metric in change['new']:
        min_input = widgets.FloatText(value=0, description='Min', step=1)
        max_input = widgets.FloatText(value=100, description='Max', step=1)
        metric_inputs[metric] = (min_input, max_input)
        inputs_vbox.children += (widgets.HBox([widgets.Label(metric), min_input, max_input]),)

# Create a VBox to hold the metric input fields
inputs_vbox = widgets.VBox([])

# Function to get the top players based on selected metrics, filters, and search criteria
def get_top_players_interactive(selected_metrics, min_minutes, top_n, player_name, team, league, nationality):
    # Start with the full dataset
    filtered_data = data

    # Filter by player name if provided
    if player_name:
        filtered_data = filtered_data[filtered_data['Name'].str.contains(player_name, case=False, na=False)]
        if filtered_data.empty:
            print(f"No player found with the name '{player_name}'.")
            return None

    # Filter by team if selected
    if team:
        filtered_data = filtered_data[filtered_data['Club'] == team]
        if filtered_data.empty:
            print(f"No players found for the team '{team}'.")
            return None

    # Filter by league if selected
    if league:
        filtered_data = filtered_data[filtered_data['Division'] == league]
        if filtered_data.empty:
            print(f"No players found in the league '{league}'.")
            return None

    # Filter by nationality if selected
    if nationality:
        filtered_data = filtered_data[filtered_data['Nationality'] == nationality]
        if filtered_data.empty:
            print(f"No players found from the nationality '{nationality}'.")
            return None

    # Further filter by minimum minutes played
    filtered_data = filtered_data[filtered_data['Minutes Played'] > min_minutes]

    # Apply min/max filters for each metric
    for metric, (min_input, max_input) in metric_inputs.items():
        filtered_data = filtered_data[
            (filtered_data[metric] >= min_input.value) & 
            (filtered_data[metric] <= max_input.value)
        ]

    if filtered_data.empty:
        print("No players match all criteria. Try adjusting your filters.")
        return None

    # Sort by the first selected metric
    if selected_metrics:
        sorted_data = filtered_data.sort_values(by=selected_metrics[0], ascending=False)
    else:
        sorted_data = filtered_data

    # Select top N players
    top_players = sorted_data.head(top_n)

    # Display only the selected metrics and additional columns
    columns_to_display = additional_columns + list(selected_metrics)
    display(top_players[columns_to_display])

    return top_players

# Connect the metric selection to input field creation
metric_select.observe(create_metric_inputs, names='value')

# Set up widgets for interactive functionality
interactive_output = widgets.interactive_output(
    get_top_players_interactive,
    {'selected_metrics': metric_select, 'min_minutes': min_minutes_slider, 'top_n': top_n_slider, 
     'player_name': player_search, 'team': team_dropdown, 'league': league_dropdown, 'nationality': nationality_dropdown}
)

# Combine all widgets into the final layout
final_widget = widgets.VBox([
    widgets.HBox([metric_select, widgets.VBox([min_minutes_slider, top_n_slider])]),
    widgets.HBox([player_search, team_dropdown, league_dropdown, nationality_dropdown]),
    inputs_vbox,
    interactive_output
])

display(final_widget)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import pandas as pd

def dynamic_clustering(data, features, n_clusters):
    # Select the specified features from the data
    data_selected = data[features]
    
    # Normalize the data using StandardScaler
    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(data_selected)
    
    # Apply KMeans clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    clusters = kmeans.fit_predict(data_scaled)
    
    # Add cluster assignments to the data
    data['Cluster'] = clusters
    
    # Apply PCA for dimensionality reduction
    pca = PCA(n_components=2)
    pca_components = pca.fit_transform(data_scaled)
    
    # Add PCA components to the data
    data['PCA1'] = pca_components[:, 0]
    data['PCA2'] = pca_components[:, 1]
    
    # Calculate cluster centroids in PCA space
    # To calculate centroids in the PCA space, we need to use the 2 PCA components
    pca_centroids = pca.transform(kmeans.cluster_centers_)
    
    # Convert centroids to a DataFrame and use PCA1 and PCA2 for the columns
    centroids = pd.DataFrame(pca_centroids, columns=['PCA1', 'PCA2'])
    
    return data, centroids
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.impute import SimpleImputer

def find_similar_players_statistics(player_name, data, n=6):
    """
    Finds the top `n` most similar players based on their statistics.
    
    Parameters:
    - player_name (str): The name of the player to search for.
    - data (DataFrame): The dataset containing player statistics.
    - n (int): The number of similar players to return (default is 5).
    
    Returns:
    - DataFrame: A DataFrame with the top `n` most similar players based on their statistics.
    """
    # Ensure the column names are strings (convert them if necessary)
    data.columns = data.columns.astype(str)
    
    # Clean column names to remove extra spaces
    data.columns = data.columns.str.strip()

    # Check if 'Name' column exists
    if 'Name' not in data.columns:
        raise KeyError("The dataset does not contain a 'Name' column. Please check the dataset structure.")
    
    # Exclude columns that are not relevant for the similarity (e.g., 'Games', 'Minutes Played')
    excluded_columns = ['Games', 'Minutes Played']
    data_relevant = data.drop(columns=excluded_columns, errors='ignore')
    
    # Retrieve the player’s stats by name
    player_stats = data_relevant[data_relevant['Name'] == player_name].iloc[0]
    
    # Ensure all columns in data are present in player_stats (fill missing columns)
    missing_columns = set(data_relevant.columns) - set(player_stats.index)
    for col in missing_columns:
        player_stats[col] = np.nan  # Fill missing columns with NaN
    
    # Convert player statistics to DataFrame for compatibility
    player_stats_df = pd.DataFrame([player_stats])
    
    # Impute missing values (e.g., by filling with the mean of the column)
    imputer = SimpleImputer(strategy='mean')  # or 'median'
    data_imputed = pd.DataFrame(imputer.fit_transform(data_relevant.select_dtypes(include=[np.number])))
    player_stats_imputed = pd.DataFrame(imputer.transform(player_stats_df[data_relevant.select_dtypes(include=[np.number]).columns]))
    
    # Standardize the data (only numeric columns)
    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(data_imputed)
    
    # Scale the input player's statistics
    player_stats_scaled = scaler.transform(player_stats_imputed)
    
    # Calculate Euclidean distances between the input player and all other players
    distances = euclidean_distances(player_stats_scaled, data_scaled)
    
    # Add distances to the dataset
    data_relevant['Distance'] = distances[0]
    
    # Sort by distance and return the top n most similar players
    similar_players = data_relevant.nsmallest(n, 'Distance')[['Name', 'Position','Height', 'Weight', 'Division', 'Club', 'Nationality', 'Wage', 'Transfer Value', 'Preferred Foot', 'Average Rating'] + list(data_relevant.select_dtypes(include=[np.number]).columns)]
    
    return similar_players

# Example usage: Input player name for comparison
player_name = input("Enter the name of the player you want to search for: ")

# Assuming `data` is your DataFrame containing player statistics
similar_players = find_similar_players_statistics(player_name, data)

print(f"Top 5 Most Similar Players to {player_name}:")
display(similar_players)
```

## Position based Statistical Similarity search

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.impute import SimpleImputer

def find_similar_players_statistics(player_name, data, n=6):
    """
    Finds the top `n` most similar players based on their statistics and positions, and also displays additional player details.
    
    Parameters:
    - player_name (str): The name of the player to search for.
    - data (DataFrame): The dataset containing player statistics and positions.
    - n (int): The number of similar players to return (default is 6).
    
    Returns:
    - DataFrame: A DataFrame with the top `n` most similar players based on their statistics and positions,
                  along with additional player details.
    """
    # Ensure the column names are strings (convert them if necessary)
    data.columns = data.columns.astype(str)
    
    # Clean column names to remove extra spaces
    data.columns = data.columns.str.strip()

    # Check if 'Name' and 'Position' columns exist
    if 'Name' not in data.columns or 'Position' not in data.columns:
        raise KeyError("The dataset does not contain 'Name' or 'Position' columns. Please check the dataset structure.")
    
    # Exclude columns that are not relevant for the similarity (e.g., 'Games', 'Minutes Played')
    excluded_columns = ['Games', 'Minutes Played']
    data_relevant = data.drop(columns=excluded_columns, errors='ignore')
    
    # Check if the player exists in the dataset
    if player_name not in data_relevant['Name'].values:
        raise ValueError(f"The player '{player_name}' does not exist in the dataset.")
    
    # Retrieve the player's stats by name
    player_stats = data_relevant[data_relevant['Name'] == player_name].iloc[0]
    
    # Ensure all columns in data are present in player_stats (fill missing columns)
    missing_columns = set(data_relevant.columns) - set(player_stats.index)
    for col in missing_columns:
        player_stats[col] = np.nan  # Fill missing columns with NaN
    
    # Get the player's position
    player_position = data_relevant.loc[data_relevant['Name'] == player_name, 'Position'].iloc[0]
    
    # Extract relevant position keywords (like "D(C)", "DM", etc.)
    def extract_positions(position):
        return set(position.split(', '))  # Positions separated by commas (e.g., "D(C), DM")

    player_positions_set = extract_positions(player_position)
    
    # Filter data to only include players with matching positions
    def match_position(player_pos):
        player_positions = extract_positions(player_pos)
        return not player_positions.isdisjoint(player_positions_set)  # Check if any positions match
    
    data_relevant['PositionMatch'] = data_relevant['Position'].apply(match_position)
    data_filtered = data_relevant[data_relevant['PositionMatch']].drop(columns=['PositionMatch'])
    
    if data_filtered.empty:
        raise ValueError("No players with matching positions found.")

    # Convert player statistics to DataFrame for compatibility
    player_stats_df = pd.DataFrame([player_stats])
    
    # Impute missing values (e.g., by filling with the mean of the column)
    imputer = SimpleImputer(strategy='mean')  # or 'median'
    data_imputed = pd.DataFrame(imputer.fit_transform(data_filtered.select_dtypes(include=[np.number])))
    player_stats_imputed = pd.DataFrame(imputer.transform(player_stats_df[data_filtered.select_dtypes(include=[np.number]).columns]))
    
    # Standardize the data (only numeric columns)
    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(data_imputed)
    
    # Scale the input player's statistics
    player_stats_scaled = scaler.transform(player_stats_imputed)
    
    # Calculate Euclidean distances between the input player and all other players
    distances = euclidean_distances(player_stats_scaled, data_scaled)
    
    # Add distances to the dataset
    data_filtered['Distance'] = distances[0]
    
    # Sort by distance and return the top n most similar players
    similar_players = data_filtered.nsmallest(n, 'Distance')[['Name', 'Distance','Height', 'Weight', 'Division', 'Club', 'Nationality', 'Wage', 'Transfer Value', 'Preferred Foot', 'Average Rating'] + list(data_filtered.select_dtypes(include=[np.number]).columns)]
    
    # Include additional player details in the output (such as Height, Weight, etc.)
    additional_columns = ['Personality']
    
    # Ensure these columns exist in the dataset
    missing_columns = [col for col in additional_columns if col not in data.columns]
    if missing_columns:
        raise KeyError(f"The dataset is missing the following columns: {', '.join(missing_columns)}")
    
    # Add the additional details to the similar players DataFrame
    similar_players_details = similar_players.merge(data[['Name'] + additional_columns], on='Name', how='left')
    
    return similar_players_details

# Example usage: Input player name for comparison
player_name = input("Enter the name of the player you want to search for: ")

# Assuming `data` is your DataFrame containing player statistics and positions
try:
    similar_players = find_similar_players_statistics(player_name, data)
    print(f"Top {len(similar_players)} Most Similar Players to {player_name}:")
    display(similar_players)
except ValueError as ve:
    print(ve)
except KeyError as ke:
    print(ke)
```

## Player Percentage breakdown
This Python script creates an interactive data analysis tool for Football Manager statistics. Below is an explanation of how the code works and what the Overall Value % represents.

### Data Loading and Preprocessing

The script starts by importing necessary libraries and loading the dataset from a CSV file. It then defines lists of metrics for different categories (attacking, creative, defensive, and goalkeeping).

### Widget Creation

The code creates several dropdown widgets using `ipywidgets`:

- **`category_dropdown`**: Allows selection of overall stats or specific categories.
- **`league_dropdown`**: Filters data by league.
- **`team_dropdown`**: Filters data by team.
- **`metric_dropdown`**: Selects a specific metric for analysis.

### Data Filtering and Analysis

The `update_display` function is the core of the analysis:

1. It filters the data based on selected league and team.
2. For the 'Overall' category, it calculates an 'Overall Value' by summing all percentile metrics.
3. For specific categories, it sums the relevant category-specific metrics.
4. It then calculates percentage values relative to the maximum in the filtered dataset.
5. Finally, it displays the top 10 players based on the selected criteria.

### Interactivity

The script uses widget observers to trigger the `update_display` function whenever a selection changes, providing real-time updates to the analysis.

## Overall Value % Explanation

The 'Overall Value %' represents a player's overall performance relative to the best performer in the current filtered dataset:

- It's calculated by summing all percentile-based metrics for a player.
- This sum is then converted to a percentage of the highest sum in the dataset.
- A player with 100% is the top performer in the current selection.
- Other players' percentages indicate how close they are to the top performer.

This percentage allows for easy comparison of players across different metrics and categories, providing a quick overview of a player's overall quality relative to their peers in the selected dataset.

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output

# Load the dataset
file_path = '/Users/JumpMan/Downloads/FM Data Analytics/Newcastle/Newcastle/Moneyball Scouting/Top 5 League Analysis/2026-2027/Summer 2026-27 Top 14 Leagues.csv'
data = pd.read_csv(file_path)

def team_league_analysis(data):
    """
    Performs team-based analysis to determine the team with the leading statistics in each league.
    
    Parameters:
    - data (DataFrame): The dataset containing player statistics, including 'Club' and 'Division'.

    Returns:
    - DataFrame: A DataFrame showing the mean stats for each team in every league.
    """
    # Check for column names in your dataset and adjust accordingly
    if 'Club' not in data.columns or 'Division' not in data.columns:
        raise ValueError("Dataset must include 'Club' and 'Division' (or similar) columns.")

    # Group by Club and Division, then calculate the mean of numeric stats
    team_stats = (
        data.groupby(['Division', 'Club'], as_index=False)
        .mean(numeric_only=True)  # Only calculate mean for numeric columns
    )
    
    # Rank teams within each league by their mean stats (sum of all means for ranking purposes)
    team_stats['Overall Score'] = team_stats.sum(axis=1, numeric_only=True)
    team_stats['Rank'] = team_stats.groupby('Division')['Overall Score'].rank(ascending=False)
    
    # Sort by Division and Rank
    team_stats = team_stats.sort_values(by=['Division', 'Rank'])
    
    return team_stats


# Perform team-based analysis
team_analysis = team_league_analysis(data)

# Create widgets
division_dropdown = widgets.Dropdown(
    options=['All'] + list(team_analysis['Division'].unique()),  # Add an 'All' option
    description='Division:',
    value='All'
)

club_dropdown = widgets.SelectMultiple(
    description='Clubs:',
    options=[],
)

top_10_button = widgets.Button(
    description="Show Top 10 Clubs",
    button_style="info"
)

output_area = widgets.Output()

# Update Club dropdown and display all teams in selected Division
def update_clubs(*args):
    with output_area:
        clear_output()
        selected_division = division_dropdown.value
        
        if selected_division == 'All':
            clubs = team_analysis['Club'].unique()
            display(team_analysis)
        else:
            filtered_data = team_analysis[team_analysis['Division'] == selected_division]
            clubs = filtered_data['Club'].unique()
            display(filtered_data)

        club_dropdown.options = clubs

# Attach the update function to the Division dropdown
division_dropdown.observe(update_clubs, names='value')
update_clubs()  # Initialize the clubs dropdown

# Display filtered results for selected clubs
def filter_data(*args):
    with output_area:
        clear_output()
        selected_division = division_dropdown.value
        selected_clubs = list(club_dropdown.value)
        
        if not selected_clubs:
            print("No clubs selected.")
            return

        filtered_data = team_analysis[
            (team_analysis['Division'] == selected_division) & 
            (team_analysis['Club'].isin(selected_clubs))
        ]
        display(filtered_data)

# Show top 10 clubs
def show_top_10_clubs(button):
    with output_area:
        clear_output()
        top_10_clubs = team_analysis.sort_values(by='Overall Score', ascending=False).head(10)
        display(top_10_clubs)

# Attach a callback to the Club dropdown and the Top 10 button
club_dropdown.observe(filter_data, names='value')
top_10_button.on_click(show_top_10_clubs)

# Display the dropdowns and output area
display(division_dropdown, club_dropdown, top_10_button, output_area)
```

## Team Similarity Statistical Search

```python
import pandas as pd
from fuzzywuzzy import process
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output

# Load the dataset
file_path = '/Users/JumpMan/Downloads/FM Data Analytics/Newcastle/Newcastle/Moneyball Scouting/Top 5 League Analysis/2026-2027/Summer 2026-27 Top 14 Leagues.csv'
data = pd.read_csv(file_path)

# Function to perform team-based analysis
def team_league_analysis(data):
    """
    Performs team-based analysis to calculate cumulative statistics for each team in each league.
    """
    # Check for required columns
    if 'Club' not in data.columns or 'Division' not in data.columns:
        raise ValueError("Dataset must include 'Club' and 'Division' columns.")

    # Group by Club and Division, then calculate the mean of numeric stats
    team_stats = (
        data.groupby(['Division', 'Club'], as_index=False)
        .mean(numeric_only=True)  # Calculate mean only for numeric columns
    )
    return team_stats

# Generate team stats
team_stats = team_league_analysis(data)

# Function to find similar teams
def find_similar_teams(team_name, team_stats, top_n=5):
    """
    Finds the top similar teams based on cumulative statistics.
    """
    if team_name not in team_stats['Club'].unique():
        # Fuzzy match team name to find the closest match
        possible_matches = process.extract(team_name, team_stats['Club'].unique(), limit=top_n)
        team_name = possible_matches[0][0]  # Choose the best match

    # Extract the searched team's stats
    searched_team_stats = team_stats[team_stats['Club'] == team_name].drop(columns=['Division', 'Club'])
    
    # Compute similarity scores with other teams
    numeric_data = team_stats.drop(columns=['Division', 'Club'])
    similarities = cosine_similarity(searched_team_stats, numeric_data)[0]

    # Add similarities to team_stats
    team_stats['Similarity'] = similarities
    
    # Get the top N similar teams
    top_similar = team_stats.sort_values(by='Similarity', ascending=False).head(top_n)
    return top_similar

# Function to find the most similar team in each league
def find_most_similar_in_league(team_name, team_stats):
    """
    Finds the most similar team to the given team in each league.
    """
    # Extract numeric stats for similarity computation
    numeric_data = team_stats.drop(columns=['Division', 'Club'])
    
    # Ensure the team exists in the data
    if team_name not in team_stats['Club'].unique():
        possible_matches = process.extract(team_name, team_stats['Club'].unique(), limit=1)
        team_name = possible_matches[0][0]
    
    searched_team_stats = team_stats[team_stats['Club'] == team_name].drop(columns=['Division', 'Club'])
    similarities = cosine_similarity(searched_team_stats, numeric_data)[0]
    
    # Add similarities to the dataset
    team_stats['Similarity'] = similarities
    
    # Find the most similar team in each league
    most_similar_teams = team_stats.loc[
        team_stats.groupby('Division')['Similarity'].idxmax()
    ][['Division', 'Club', 'Similarity']]
    
    return most_similar_teams

# Create widgets for searching and displaying results
team_search_box = widgets.Text(
    placeholder='Enter team name',
    description='Team:',
    style={'description_width': 'initial'}
)
top_n_slider = widgets.IntSlider(value=5, min=1, max=10, step=1, description='Top N Similar:')
search_button = widgets.Button(description='Search')
output_area = widgets.Output()

# Function to handle the search
def search_team_results(b):
    with output_area:
        clear_output()
        team_name = team_search_box.value
        top_n = top_n_slider.value
        
        if team_name:
            # Find top similar teams
            try:
                similar_teams = find_similar_teams(team_name, team_stats, top_n)
                print(f"Top {top_n} Teams Similar to '{team_name}':")
                display(similar_teams[['Division', 'Club', 'Similarity']])
                
                # Find most similar team in each league
                most_similar = find_most_similar_in_league(team_name, team_stats)
                print(f"\nMost Similar Team in Each League to '{team_name}':")
                display(most_similar)
            except Exception as e:
                print(f"Error: {e}")
        else:
            print("Please enter a team name.")

# Attach search button click event
search_button.on_click(search_team_results)

# Display widgets
display(team_search_box, top_n_slider, search_button, output_area)
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import euclidean, correlation
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output
from fuzzywuzzy import process
from sklearn.preprocessing import StandardScaler

# Load the dataset
file_path = '/Users/JumpMan/Downloads/FM Data Analytics/Newcastle/Newcastle/Moneyball Scouting/Top 5 League Analysis/2026-2027/Summer 2026-27 Top 14 Leagues.csv'
data = pd.read_csv(file_path)

# Define metric categories
metric_categories = {
    "Attacking": [
        'Goals (percentile)', 'Goals/90 (percentile)', 'Minutes/Goal (percentile)', 'xG (percentile)',
        'xG/90 (percentile)', 'xG/Shot (percentile)', 'Non-pen Goals (percentile)', 'Non-pen Goals/90 (percentile)',
        'Non-pen Goals/Shot (percentile)', 'Non-pen xG (percentile)', 'Non-pen xG/90 (percentile)', 'Shots (percentile)',
        'Shots/90 (percentile)', 'Shots on Target (percentile)', 'Shots on Target/90 (percentile)', 'Shots on Target % (percentile)',
        'Shots Outside Box/90 (percentile)', 'Goal Contributions (percentile)', 'Goal Contributions/90 (percentile)',
        'Goals Outside Box (percentile)', 'Goals Outside Box/90 (percentile)', 'Conversion % (percentile)', 
        'Penalties Taken (percentile)', 'Penalties Scored (percentile)', 'Pens Scored % (percentile)'
],
    "Creative": [
        'Assists (percentile)', 'Assists/90 (percentile)', 'xA (percentile)', 'xA/90 (percentile)', 'xA Overperformance (percentile)',
        'Key Passes (percentile)', 'Key Passes/90 (percentile)', 'Key Pass % (percentile)', 'Open Play Key Passes (percentile)',
        'Chances Created (percentile)', 'Chances Created/90 (percentile)', 'Clear Cut Chances Created (percentile)',
        'Clear Cut Chances Created/90 (percentile)', 'Progressive Passes/90 (percentile)', 'Progressive Passes (percentile)',
        'Pass Completion % (percentile)', 'Crosses Attempted (percentile)', 'Crosses Attempted/90 (percentile)',
        'Crosses Completed (percentile)', 'Crosses Completed/90 (percentile)', 'Passes Attempted (percentile)', 
        'Passes Attempted/90 (percentile)'
    ],
    "Defensive": [
        'Tackles Attempted (percentile)', 'Tackles Attempted/90 (percentile)', 'Tackles Completed (percentile)',
        'Tackles Completed/90 (percentile)', 'Interceptions (percentile)', 'Interceptions/90 (percentile)', 
        'Clearances (percentile)', 'Clearances/90 (percentile)', 'Blocks (percentile)', 'Blocks/90 (percentile)', 
        'Key Tackles (percentile)', 'Key Tackles/90 (percentile)', 'Tackle Completion % (percentile)', 
        'Defensive Actions/90 (percentile)', 'Fouls Committed/90 (percentile)', 'Possession Won/90 (percentile)', 
        'Duels Win % (percentile)', 'Headers Won % (percentile)'
    
    ],
    "Goalkeeping": [
        'Total Saves (percentile)', 'Saves/90 (percentile)', 'Save % (percentile)',
        'xSave % (percentile)', 'xSave % Overperformance (percentile)', 
        'Saves Held (percentile)', 'Saves Held/90 (percentile)', 'Saves Held Ratio (percentile)',
        'Saves Tipped (percentile)', 'Saves Tipped/90 (percentile)'

    ]
}

# Normalize data
def normalize_data(data, metrics):
    """
    Normalizes the data for the given metrics using StandardScaler.
    """
    scaler = StandardScaler()
    data[metrics] = scaler.fit_transform(data[metrics])
    return data

# Function to calculate weighted similarity
def calculate_similarity(team1, team2, data, metrics, weights=None):
    """
    Calculates similarity between two teams using cosine, Euclidean, and correlation metrics.
    """
    # Normalize metrics
    normalized_data = normalize_data(data, metrics)
    
    # Fuzzy match team names
    team1_name = process.extractOne(team1, normalized_data['Club'].unique())[0]
    team2_name = process.extractOne(team2, normalized_data['Club'].unique())[0]
    
    # Extract metrics for the teams
    team1_metrics = normalized_data[normalized_data['Club'] == team1_name][metrics].values.flatten()
    team2_metrics = normalized_data[normalized_data['Club'] == team2_name][metrics].values.flatten()
    
    # Apply weights
    if weights:
        team1_metrics = team1_metrics * weights
        team2_metrics = team2_metrics * weights
    
    # Similarity metrics
    cosine_sim = cosine_similarity([team1_metrics], [team2_metrics])[0][0]
    euclidean_sim = 1 / (1 + euclidean(team1_metrics, team2_metrics))  # Inverse to make it a similarity measure
    correlation_sim = 1 - correlation(team1_metrics, team2_metrics)
    
    # Aggregate similarity
    overall_similarity = (cosine_sim + euclidean_sim + correlation_sim) / 3
    
    # Metric-wise similarity
    metric_similarities = {}
    for i, metric in enumerate(metrics):
        metric_similarities[metric] = cosine_similarity([[team1_metrics[i]]], [[team2_metrics[i]]])[0][0] * 100
    
    return round(overall_similarity * 100, 2), metric_similarities

# Team-based analysis
def team_league_analysis(data):
    """
    Aggregates data by team.
    """
    if 'Club' not in data.columns or 'Division' not in data.columns:
        raise ValueError("Dataset must include 'Club' and 'Division' columns.")
    return data.groupby(['Division', 'Club'], as_index=False).mean(numeric_only=True)

# Generate team stats
team_stats = team_league_analysis(data)

# Interactive Widgets
team1_box = widgets.Text(placeholder='Enter Team 1', description='Team 1:', style={'description_width': 'initial'})
team2_box = widgets.Text(placeholder='Enter Team 2', description='Team 2:', style={'description_width': 'initial'})
category_dropdown = widgets.Dropdown(options=metric_categories.keys(), description='Category:')
weights_slider = widgets.FloatSlider(min=0, max=2, step=0.1, value=1, description='Weight:', style={'description_width': 'initial'})
search_button = widgets.Button(description='Compare')
output_area = widgets.Output()

# Compare teams function
def compare_teams(b):
    with output_area:
        clear_output()
        team1 = team1_box.value
        team2 = team2_box.value
        category = category_dropdown.value
        
        if not team1 or not team2:
            print("Please enter both team names.")
            return
        
        # Get metrics and weights
        metrics = metric_categories[category]
        weights = [weights_slider.value] * len(metrics)
        
        # Calculate similarity
        overall_similarity, metric_similarities = calculate_similarity(team1, team2, team_stats, metrics, weights)
        
        print(f"Overall Similarity between {team1} and {team2}: {overall_similarity}%")
        print("\nMetric-wise Similarities:")
        for metric, similarity in metric_similarities.items():
            print(f"{metric}: {similarity:.2f}%")

# Attach button event
search_button.on_click(compare_teams)

# Display widgets
display(team1_box, team2_box, category_dropdown, weights_slider, search_button, output_area)
```

```python
import locale
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

# Set locale for GBP formatting
locale.setlocale(locale.LC_ALL, 'en_GB.UTF-8')

# Define the metric categories
metric_categories = {
    "Goalkeeping": {
        "metrics": [
            'Total Saves (percentile)', 'Saves/90 (percentile)', 'Save % (percentile)', 'xSave % (percentile)', 
            'xSave % Overperformance (percentile)', 'Saves Held (percentile)', 'Saves Held/90 (percentile)', 
            'Saves Held Ratio (percentile)', 'Saves Tipped (percentile)', 'Saves Tipped/90 (percentile)'
        ],
        "positions": ["GK"]
    },
    "Defensive": {
        "metrics": [
            'Tackles Attempted (percentile)', 'Tackles Attempted/90 (percentile)', 'Tackles Completed (percentile)',
            'Tackles Completed/90 (percentile)', 'Interceptions (percentile)', 'Interceptions/90 (percentile)', 
            'Clearances (percentile)', 'Clearances/90 (percentile)', 'Blocks (percentile)', 'Blocks/90 (percentile)', 
            'Key Tackles (percentile)', 'Key Tackles/90 (percentile)', 'Tackle Completion % (percentile)', 
            'Defensive Actions/90 (percentile)', 'Fouls Committed/90 (percentile)', 'Possession Won/90 (percentile)', 
            'Duels Win % (percentile)', 'Headers Won % (percentile)'
        ],
        "positions": [
            "D (C)", "D (RL)", "WB/M/AM (RL)", "WB/M/AM (R)", "WB/M (RL)", "WB/M (L)", "WB (RL)", 
            "WB (R)", "DM", "WB/AM (RL)", "WB/AM (R)", "WB/AM (L)"
        ]
    },
    "Creative": {
        "metrics": [
            'Assists (percentile)', 'Assists/90 (percentile)', 'xA (percentile)', 'xA/90 (percentile)', 'xA Overperformance (percentile)',
            'Key Passes (percentile)', 'Key Passes/90 (percentile)', 'Key Pass % (percentile)', 'Open Play Key Passes (percentile)',
            'Chances Created (percentile)', 'Chances Created/90 (percentile)', 'Clear Cut Chances Created (percentile)',
            'Clear Cut Chances Created/90 (percentile)', 'Progressive Passes/90 (percentile)', 'Progressive Passes (percentile)',
            'Pass Completion % (percentile)', 'Crosses Attempted (percentile)', 'Crosses Attempted/90 (percentile)',
            'Crosses Completed (percentile)', 'Crosses Completed/90 (percentile)', 'Passes Attempted (percentile)', 
            'Passes Attempted/90 (percentile)'
        ],
        "positions": [
            "AM (RLC)", "AM (RL)", "AM (RC)", "AM (R)", "AM (LC)", "AM (L)", "AM (C)", "WB/M (RLC)", 
            "WB/M (R)", "WB/M (RL)"
        ]
    },
    "Attacking": {
        "metrics": [
            'Goals (percentile)', 'Goals/90 (percentile)', 'Minutes/Goal (percentile)', 'xG (percentile)',
            'xG/90 (percentile)', 'xG/Shot (percentile)', 'Non-pen Goals (percentile)', 'Non-pen Goals/90 (percentile)',
            'Non-pen Goals/Shot (percentile)', 'Non-pen xG (percentile)', 'Non-pen xG/90 (percentile)', 'Shots (percentile)',
            'Shots/90 (percentile)', 'Shots on Target (percentile)', 'Shots on Target/90 (percentile)', 'Shots on Target % (percentile)',
            'Shots Outside Box/90 (percentile)', 'Goal Contributions (percentile)', 'Goal Contributions/90 (percentile)',
            'Goals Outside Box (percentile)', 'Goals Outside Box/90 (percentile)', 'Conversion % (percentile)', 
            'Penalties Taken (percentile)', 'Penalties Scored (percentile)', 'Pens Scored % (percentile)'    
        ],
        "positions": ["ST", "AM", "WB/AM (RL)", "WB/AM (R)", "WB/AM (L)"]
    }
}

# Preprocess Wage column to remove non-numeric characters and convert to float
def preprocess_wage_column(data):
    data['Wage'] = (
        data['Wage']
        .replace(r'[^\d.]', '', regex=True)  # Remove all non-numeric characters except '.'
        .astype(float)  # Convert to float
    )
    return data

# Normalize data for relevant metrics
def normalize_data(data, metrics):
    scaler = MinMaxScaler()
    valid_metrics = [metric for metric in metrics if metric in data.columns]
    if valid_metrics:
        data[valid_metrics] = scaler.fit_transform(data[valid_metrics])
    return data

# Calculate a combined score based on the player metrics
def calculate_combined_score(player, metric_categories):
    combined_score = 0
    for category, details in metric_categories.items():
        for metric in details['metrics']:
            combined_score += player.get(metric, 0)
    return combined_score

# Scale combined scores to a 0-100 scale
def scale_combined_scores(data):
    scaler = MinMaxScaler(feature_range=(0, 100))
    data['Combined Score'] = scaler.fit_transform(data[['Combined Score']])
    return data

# Identify underperforming players based on average rating and wage
def identify_underperforming_players(data, club_name):
    def calculate_underperformance(player):
        if player['Minutes Played'] < 500 or player['Average Rating'] < 6.00:
            return float('nan')  # Return NaN for players who don't meet the criteria
        
        rating_score = player['Average Rating']
        wage_score = player['Wage']
        
        # Normalize rating and wage for a fair comparison
        rating_normalized = 100 - (rating_score * 100 / 10)
        wage_normalized = 100 - (wage_score / 500_000 * 100)

        # Weighted underperformance score
        underperformance_score = (rating_normalized * 0.6) + (wage_normalized * 0.4)
        return underperformance_score

    club_data = data[data['Club'] == club_name]
    club_data['Underperformance'] = club_data.apply(calculate_underperformance, axis=1)
    
    return club_data.nlargest(5, 'Underperformance')

# Calculate similarity between player and recommended player based on metrics
def calculate_similarity(player, recommended_player, metric_categories):
    metrics = [metric for category in metric_categories.values() for metric in category['metrics']]
    
    player_metrics = player[metrics].values
    recommended_player_metrics = recommended_player[metrics].values
    
    difference = player_metrics - recommended_player_metrics
    similarity = 100 - (abs(difference).sum() / len(metrics)) * 100
    
    return max(0, min(similarity, 100))

# Recommend upgrades for a player based on position, rating, and budget range
def recommend_upgrades_for_player(player, data, budget_range, metric_categories):
    position = player['Position']
    average_rating = player['Average Rating']
    relative_positions = [position]  # Extend this logic for similar positions

    potential_upgrades = data[
        (data['Club'] != player['Club']) &
        (data['Position'].isin(relative_positions)) &
        (data['Transfer Value'] >= budget_range[0]) &
        (data['Transfer Value'] <= budget_range[1]) &
        (data['Average Rating'] > average_rating) &
        (data['Minutes Played'] >= 1000) &
        (data['Average Rating'] >= 7.00)
    ].copy()

    if potential_upgrades.empty:
        return pd.DataFrame()

    potential_upgrades['Combined Score'] = potential_upgrades.apply(
        lambda row: calculate_combined_score(row, metric_categories), axis=1
    )
    
    if potential_upgrades['Combined Score'].isnull().all():
        return pd.DataFrame()

    potential_upgrades['Combined Score'] = MinMaxScaler(feature_range=(0, 100)).fit_transform(
        potential_upgrades[['Combined Score']]
    )
    
    potential_upgrades['Similarity (%)'] = potential_upgrades.apply(
        lambda row: calculate_similarity(player, row, metric_categories), axis=1
    )
    
    potential_upgrades['Transfer Value'] = potential_upgrades['Transfer Value'].apply(
        lambda x: locale.currency(x, grouping=True)
    )
    potential_upgrades['Wage'] = potential_upgrades['Wage'].apply(
        lambda x: locale.currency(x, grouping=True)
    )
    
    return potential_upgrades.nlargest(5, 'Combined Score')[['Name', 'Club', 'Age', 'Position', 'Transfer Value', 'Wage', 'Average Rating', 'Combined Score', 'Similarity (%)']]

# Main function to recommend upgrades for underperforming players
def recommend_upgrades(club_name, min_budget, max_budget, data, metric_categories):
    budget_range = (min_budget * 1_000_000, max_budget * 1_000_000)
    
    underperforming_players = identify_underperforming_players(data, club_name)
    
    recommendations = {}
    for _, player in underperforming_players.iterrows():
        recommendations[player['Name']] = recommend_upgrades_for_player(player, data, budget_range, metric_categories)
    
    return underperforming_players, recommendations

# Event handler for button click in Jupyter widgets
def on_recommend_click(b):
    with output_area:
        clear_output()
        club_name = club_box.value
        min_budget = min_budget_box.value
        max_budget = max_budget_box.value
        
        underperforming_players, recommendations = recommend_upgrades(club_name, min_budget, max_budget, data, metric_categories)
        
        print("<h3 style='color: blue;'>Top 5 Underperforming Players:</h3>")
        if not underperforming_players.empty:
            underperforming_players['Wage'] = underperforming_players['Wage'].apply(lambda x: locale.currency(x, grouping=True))
            underperforming_players['Transfer Value'] = underperforming_players['Transfer Value'].apply(lambda x: locale.currency(x, grouping=True))
            display(HTML(underperforming_players.to_html(index=False)))
        else:
            print("<p>No underperforming players found.</p>")
        
        for player, upgrades in recommendations.items():
            print(f"<h4 style='color: green;'>Recommendations for {player}:</h4>")
            if not upgrades.empty:
                display(HTML(upgrades.to_html(index=False)))
            else:
                print("<p>No upgrades found within the budget range.</p>")

# Preprocess the data and clean the Wage column
data = preprocess_wage_column(data)

# Create widgets for user interaction
club_box = widgets.Dropdown(
    options=data['Club'].unique(),
    description='Club:',
    disabled=False
)

min_budget_box = widgets.IntText(
    value=50,  # Minimum value for budget input
    description='Min Budget (€M):',
    disabled=False
)

max_budget_box = widgets.IntText(
    value=100,  # Maximum value for budget input
    description='Max Budget (€M):',
    disabled=False
)

recommend_button = widgets.Button(
    description='Recommend Upgrades',
    disabled=False
)

output_area = widgets.Output()

# Bind button to event handler
recommend_button.on_click(on_recommend_click)

# Display widgets
display(club_box, min_budget_box, max_budget_box, recommend_button, output_area)
```
